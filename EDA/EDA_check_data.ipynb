```python
result = df_customer.groupby('customer_unique_id')['customer_id'].nunique().reset_index()
result.columns = ['customer_unique_id', 'count_customer_id']
result.head(5)
```

```python
# 구매 횟수별 고객 수 계산
purchase_count = result.groupby('count_customer_id').size().reset_index(name='count_customer_unique_id')

# 바차트 그리기
plt.bar(purchase_count['count_customer_id'], purchase_count['count_customer_unique_id'])
plt.xlabel('Purchase Count')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# 구매 횟수가 3회 이상인 그룹 처리
purchase_count.loc[purchase_count['count_customer_id'] >= 3, 'count_customer_id'] = '3 or More'
purchase_count = purchase_count.groupby('count_customer_id').sum().reset_index()

# 구매 횟수별 비율 계산
total_customers = purchase_count['count_customer_unique_id'].sum()
purchase_count['ratio'] = purchase_count['count_customer_unique_id'] / total_customers

# 파이차트 그리기
labels = ['1 Purchase', '2 Purchases', '3 or More Purchases']
plt.pie(purchase_count['ratio'], labels=labels, autopct='%1.1f%%')
plt.title('Purchase Count Ratio')
plt.show()
```
